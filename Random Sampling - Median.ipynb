```python
#Importing the Libraries

import numpy as np
import pandas as pd
import random
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
import math
import random
import statistics as stat
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#loading the dataset

pid = pd.read_csv('diabetes.csv')
pid.head()
```

```python
pid_age = pid['Age']
pid_age.describe()
```

```python
pid_age.dtype
```

#### Population Median

```python
pop_values, pop_frequencies = np.unique(pid_age, return_counts=True)

plt.figure(figsize=(15,8))
sns.barplot(x=pop_values,y=pop_frequencies,color = 'brown')

# Plot formatting
plt.title('Frequency chart of Population - Age distribution',fontsize='14')
sns.despine()
plt.xlabel('Ages',fontsize='12')
plt.ylabel('Frequency',fontsize='12') 

print(f'median of the population = {stat.median(pid.Age):.3f}')
```

## For sample size - 30
#### 1. Random sampling with Sample size = 30 and Sample count = 100

```python
np.random.seed(100) #setting the seed to an arbitary value

sample_count = 100
sample_size = 30

median_sample_values1  = []

for i in range(sample_count):
    samples1 = (np.random.choice(pid_age,sample_size))
    median_sample_values1.append(int(stat.median(samples1)))
    
values1, frequencies1 = np.unique(median_sample_values1, return_counts=True)

plt.figure(figsize=(8,5))
sns.barplot(x=values1,y=frequencies1,color = 'brown')

# Plot formatting
plt.title('Sampling distribution of median, sample size - 30, sample count-100',fontsize='14')
sns.despine()
plt.xlabel('Ages',fontsize='12')
plt.ylabel('Frequency',fontsize='12') 

print(f'Median of the Sample: {stat.median(median_sample_values1):.3f}')
```

#### Sampling distributions of sample sizes -10,30,100 for Sample count - 500

```python
np.random.seed(100)

array1 = []
array2 = []
array3 = []
n = 500
for i in range(1,n):
    array1.append(stat.median(np.random.choice(pid_age,10,replace= True)))
    array2.append(stat.median(np.random.choice(pid_age,30,replace= True)))
    array3.append(stat.median(np.random.choice(pid_age,100,replace= True)))
    
#print(array)
fig , (ax1,ax2,ax3) = plt.subplots(nrows=1, ncols=3,figsize=(18,5))
#plt.figure()

#plt.subplot(1)
sns.histplot(array1,bins=10, ax=ax1, color = 'teal',kde=True)
ax1.set_xlabel('Ages')
ax1.set_ylabel('frequency')
ax1.set_title('Sample size = 10')
ax1.axvline(x=np.median(array1),color='r') 

#ax2.subplot(1)
sns.histplot(array2, bins=10, ax=ax2, color = 'g',kde=True)
ax2.set_xlabel('Ages')
ax2.set_ylabel('frequency')
ax2.set_title('Sample size = 30')
ax2.axvline(x=np.median(array2),color='r') 

#ax3.subplot(3)
sns.histplot(array3,bins = 10,ax=ax3, kde=True)
ax3.set_xlabel('Ages')
ax3.set_ylabel('frequency')
ax3.set_title('Sample size = 100')
ax3.axvline(x=np.median(array3),color='r') 

print(f'Mean of the Sample size - 10:  {np.median(array1):.3f}')
print(f'Mean of the Sample size - 30:  {np.median(array2):.3f}')
print(f'Mean of the Sample size - 100: {np.median(array3):.3f}')
```

1. The Median of the population  = 29.000
2. The overall estimates of the random sampling of the median for the sample size - 10  =  29.250
3. The overall estimates of the random sampling of the median for the sample size - 30  =  29.125
4. The overall estimates of the random sampling of the median for the sample size -100  =  29.000
5. From the above histograms, can see by increasing the sample size, the sample median becomes more close to the Population data median
