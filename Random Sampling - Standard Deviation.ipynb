```python
#Importing the Libraries

import numpy as np
import pandas as pd
import random
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
import math
import random
import statistics as stat
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#loading the dataset

pid = pd.read_csv('diabetes.csv')
pid.head()
```

```python
pid_age = pid['Age']
pid_age.describe()
```

```python
pid_age.dtype
```

#### Population Standard deviation

```python
pop_values, pop_frequencies = np.unique(pid_age, return_counts=True)

plt.figure(figsize=(15,8))
sns.barplot(x=pop_values,y=pop_frequencies,color = 'brown')

# Plot formatting
plt.title('Frequency chart of Population - Age distribution',fontsize='14')
sns.despine()
plt.xlabel('Ages',fontsize='12')
plt.ylabel('Frequency',fontsize='12') 

print(f'Standard deviation of the population  = {np.std(pid.Age):.3f}')
```

## For sample size - 30
#### 1. Random sampling with Sample size = 30 and Sample count = 100

```python
np.random.seed(100) #setting the seed to an arbitary value

sample_count = 100
sample_size = 30

std_sample_values1  = []

for i in range(sample_count):
    samples1 = (np.random.choice(pid_age,sample_size))
    std_sample_values1.append(int(np.std(samples1)))
    
values1, frequencies1 = np.unique(std_sample_values1, return_counts=True)

plt.figure(figsize=(8,5))
sns.barplot(x=values1,y=frequencies1,color = 'brown')

# Plot formatting
plt.title('Sampling distribution of standard deviation,sample size - 30,sample count-100',fontsize='14')
sns.despine()
plt.xlabel('Ages',fontsize='12')
plt.ylabel('Frequency',fontsize='12') 

print(f'Standard deviation of the Sample: {np.std(std_sample_values1):.3f}')
```

#### 2. Random sampling with Sample size = 30 and Sample count = 1000

```python
np.random.seed(100) #setting the seed to an arbitary value

sample_count = 1000
sample_size = 30

std_sample_values2  = []

for i in range(sample_count):
    samples2 = (np.random.choice(pid_age,sample_size))
    std_sample_values2.append(int(stat.pstdev(samples2)))
    
values2, frequencies2 = np.unique(std_sample_values2, return_counts=True)

plt.figure(figsize=(8,5))
sns.barplot(x=values2,y=frequencies2,color = 'brown')

# Plot formatting
plt.title('Sampling distribution of standard deviation, sample size - 30,sample count-1000',fontsize='14')
sns.despine()
plt.xlabel('Ages',fontsize='12')
plt.ylabel('Frequency',fontsize='12') 

print(f'Standard deviation of the Sample: {stat.stdev(std_sample_values2):.3f}')
```

#### 3. Random sampling with Sample size = 30 and Sample count = 10000

```python
np.random.seed(100) #setting the seed to an arbitary value

sample_count = 10000
sample_size = 30

std_sample_values3  = []

for i in range(sample_count):
    samples3 = (np.random.choice(pid_age,sample_size))
    std_sample_values3.append(int(stat.pstdev(samples3)))
    
values3, frequencies3 = np.unique(std_sample_values3, return_counts=True)

plt.figure(figsize=(8,5))
sns.barplot(x=values3,y=frequencies3,color = 'brown')

# Plot formatting
plt.title('Sampling distribution of standard deviation, sample size - 30,sample count-10000',fontsize='14')
sns.despine()
plt.xlabel('Ages',fontsize='12')
plt.ylabel('Frequency',fontsize='12') 

print(f'Standard deviation of the Sample: {stat.pstdev(std_sample_values3):.3f}')
```

```python
#Overall estimates of the Sample Standard deviation for sample size 30

print(f'overall estimate of Standard deviation for sample size 30 is:\
     {(np.std(std_sample_values1) + stat.stdev(std_sample_values2) + stat.pstdev(std_sample_values3))/3:.3f}')
```

#### Sampling distributions of sample sizes -10,30,100 for Sample count - 500

```python
np.random.seed(100)

array1 = []
array2 = []
array3 = []
n = 500
for i in range(1,n):
    array1.append(stat.pstdev(np.random.choice(pid_age,10,replace= True)))
    array2.append(stat.pstdev(np.random.choice(pid_age,30,replace= True)))
    array3.append(stat.pstdev(np.random.choice(pid_age,100,replace= True)))
    
#print(array)
fig , (ax1,ax2,ax3) = plt.subplots(nrows=1, ncols=3,figsize=(18,5))
#plt.figure()

#plt.subplot(1)
sns.histplot(array1,bins=10, ax=ax1, color = 'teal',kde=True)
ax1.set_xlabel('Ages')
ax1.set_ylabel('frequency')
ax1.set_title('Sample size = 10')

#ax2.subplot(1)
sns.histplot(array2, bins=10, ax=ax2, color = 'g',kde=True)
ax2.set_xlabel('Ages')
ax2.set_ylabel('frequency')
ax2.set_title('Sample size = 30')

#ax3.subplot(3)
sns.histplot(array3,bins = 10,ax=ax3, kde=True)
ax3.set_xlabel('Ages')
ax3.set_ylabel('frequency')
ax3.set_title('Sample size = 100')

print(f'Mean of the Sample size - 10:  {np.std(array1):.3f}')
print(f'Mean of the Sample size - 30:  {np.std(array2):.3f}')
print(f'Mean of the Sample size - 100: {np.std(array3):.3f}')
```

4. The overall estimates of the random sampling of the standard deviation for the sample size -100 =  1.002
